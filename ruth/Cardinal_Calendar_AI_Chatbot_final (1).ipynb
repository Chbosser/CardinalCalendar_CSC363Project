```python
import glob
from langchain_community.document_loaders import DirectoryLoader, PyPDFLoader

loader = DirectoryLoader(
    r"C:\Users\ruthe\OneDrive\Documents\Cardinal Calendar",  # your folder
    glob="*.pdf",             # load all PDFs
    loader_cls=PyPDFLoader    # use the PDF loader
)

docs_pdf = loader.load()

print(f"Loaded {len(docs_pdf)} pages from all PDFs")

```

```python
documents = docs_web + docs_pdf
```

```python
%pip uninstall torch -y
%pip uninstall sentence-transformers -y
%pip install torch "sentence-transformers==3.0.0"
```

```python
# Splitting documents
from langchain_text_splitters import RecursiveCharacterTextSplitter
doc_splitter = RecursiveCharacterTextSplitter(
    chunk_size=10000,
    chunk_overlap=200
)

chunks = doc_splitter.split_documents(documents)
```

```python
%pip uninstall torch sentence_transformers
%pip install torch "sentence_transformers == 3.0.0"
```

```python
doc_emb = SentenceTransformersDocumentEmbedder(model="sentence-transformers/all-MiniLM-L6-v2")
txt_emb = SentenceTransformersTextEmbedder(model="sentence-transformers/all-MiniLM-L6-v2")
doc_emb.warm_up()
txt_emb.warm_up()

# Create a new InMemoryDocumentStore for the chunks
chunk_store = InMemoryDocumentStore()

# Convert LangChain Documents to Haystack Documents
haystack_documents = []
for doc in chunks:
    haystack_documents.append(Document(content=doc.page_content, meta=doc.metadata))

chunk_store.write_documents(haystack_documents)

embedded = doc_emb.run(documents=chunk_store.filter_documents())["documents"]
chunk_store.write_documents(embedded, policy="overwrite")
```

```python
# 3  Retrievers
bm25  = InMemoryBM25Retriever(document_store=chunk_store, top_k=3)
dense = InMemoryEmbeddingRetriever(document_store=chunk_store, top_k=3)
```

```python
# 4  Retrieval-only pipeline
pipe = Pipeline()

# Re-instantiate components to avoid PipelineError
bm25  = InMemoryBM25Retriever(document_store=chunk_store, top_k=3)
dense = InMemoryEmbeddingRetriever(document_store=chunk_store, top_k=3)
txt_emb = SentenceTransformersTextEmbedder(model="sentence-transformers/all-MiniLM-L6-v2")

pipe.add_component("bm25", bm25)
pipe.add_component("dense", dense)
pipe.add_component("txt_emb", txt_emb)
pipe.connect("txt_emb.embedding", "dense.query_embedding")
```

```python
%pip uninstall pydantic -y
%pip uninstall langchain-core -y
%pip uninstall langchain-community -y
%pip uninstall unstructured -y
```

```python
q = "What classes should I take in year 1?"
res = pipe.run({"bm25": {"query": q}, "txt_emb": {"text": q}})
docs = res["bm25"]["documents"] + res["dense"]["documents"]
context = "\n".join({d.content for d in docs})
```
